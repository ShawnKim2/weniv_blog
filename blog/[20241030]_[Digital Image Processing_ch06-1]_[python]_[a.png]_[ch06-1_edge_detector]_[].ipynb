```python
import cv2
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/gdrive')
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = '/content/gdrive/MyDrive/Colab Notebooks/Images/lena256.jpg'

origin_img = cv2.imread(img_path)
RGB_img = cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)

cv2_imshow(origin_img)
```

```python
g_img = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
cv2_imshow(g_img)
```

```python
gx = np.array([[-1, 0], [0, 1]],dtype = int)
gy = np.array([[0, 1], [-1, 0]],dtype = int)
```

```python
x = cv2.filter2D(g_img, -1, gx)
y = cv2.filter2D(g_img, -1, gy)
```

```python
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)

out_img = cv2.addWeighted(absX, 0.5, absY, 0.5, 0)
```

```python
cv2_imshow(out_img)
```
